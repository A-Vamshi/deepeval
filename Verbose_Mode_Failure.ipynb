```python
from deepeval.metrics import BaseMetric, AnswerRelevancyMetric
from typing import List
from deepeval.utils import prettify_list

def construct_verbose_logs(metric: BaseMetric, steps: List[str]) -> str:
    verbose_logs = ""
    for i in range(len(steps) - 1):
        verbose_logs += steps[i]

        # don't add new line for penultimate step
        if i < len(steps) - 2:
            verbose_logs += " \n \n"

    if metric.verbose_mode:
        # only print reason and score for deepeval
        print_verbose_logs(metric.__name__, verbose_logs + f"\n \n{steps[-1]}")

    return verbose_logs

def print_verbose_logs(metric: str, logs: str):
    print("*" * 50)
    print(f"{metric} Verbose Logs")
    print("*" * 50)
    print("")
    print(logs)
    print("")
    print("=" * 70)


metric = AnswerRelevancyMetric(verbose_mode=True)

statements = [
    "It depends.",
    "Some might consider the cat.",
    "Others might argue the dog."
]

verdicts = [
    {"verdict": "idk", "reason": None},
    {"verdict": "yes", "reason": None},
    {"verdict": "yes", "reason": None},
]

score = 1
reason = """The score is 1.00 because the output accurately identifies 'the cat' as the one who ran up the tree based
on the input statement, without any irrelevant information. Well done!"""

steps = [
    f"Statements:\n{prettify_list(statements)}",
    f"Verdicts:\n{prettify_list(verdicts)}",
    f"Score: {score}\nReason: {reason}",
]

metric.verbose_logs = construct_verbose_logs(metric, steps)
```

```python
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.test_case import LLMTestCaseParams
from deepeval.test_case import LLMTestCase


test_case = LLMTestCase(
    input="The dog chased the cat up the tree, who ran up the tree?",
    actual_output="It depends, some might consider the cat, while others might argue the dog.",
    expected_output="The cat."
)

metric = AnswerRelevancyMetric(verbose_mode=True, async_mode=False)

metric.measure(test_case)
print(metric.score)
print(metric.reason)
```
